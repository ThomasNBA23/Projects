## ðŸ“ˆ Poisson-Based Season Simulation

We simulate the Premier League season using a Poisson distribution model for goal scoring.

The model assumes the number of goals scored by each team in a match follows a Poisson distribution, with rates estimated from historical averages:

Average goals scored and conceded at home,
Average goals scored and conceded away.
We run 1,000 Monte Carlo simulations to estimate the distribution of final rankings. This provides insights such as:

Probabilities of winning the league or finishing top 4,
Median positions,
Range of uncertainty for each team.
This model is simplistic yet surprisingly effective for predicting football outcomes, especially over a full season.
